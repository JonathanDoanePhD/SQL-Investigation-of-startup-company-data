```python
#Let's find the column names in the StartupData database, which uses the "startup data.csv" found here:
# https://www.kaggle.com/manishkc06/startup-success-prediction

query_columns = """
SELECT TOP 5 *
FROM StartupData.dbo.startup_data
"""
```

```python
%sql $query_columns
```

```python
#Let's learn more about our data

import pandas as pd

query_full = """
SELECT *
FROM StartupData.dbo.startup_data
"""

Full = %sql $query_full

FullDF = pd.DataFrame(Full)

print("____________________________________________________________", "\n")

print("#rows =", len(FullDF), "", "#cols =", len(FullDF.columns), "\n")

FullDF.head()
```

```python
#The success of a startup is indicated in the status columns as "acquired" (as opposed to "closed")

query_success = """
SELECT *
FROM StartupData.dbo.startup_data
WHERE status = 'acquired'
"""

query_fail = """
SELECT *
FROM StartupData.dbo.startup_data
WHERE status = 'closed'
"""


#Running the SQL queries will result in large tables; let's just save the results as follows

Success = %sql $query_success

Fail = %sql $query_fail


SuccessDF = pd.DataFrame(Success)

FailDF = pd.DataFrame(Fail)


print("____________________________________________________________", "\n")

print("We have accounted for", len(SuccessDF),"+",len(FailDF),"=",len(FailDF)+len(SuccessDF), "of", len(FullDF), "rows.", "\n")

print(len(FullDF),"-",len(FailDF)+len(SuccessDF),"=",len(FullDF)-(len(FailDF)+len(SuccessDF)), 'rows must have "status" reported as something else;', "let's investigate!") 
```

```python
query_unknown = """
SELECT *
FROM StartupData.dbo.startup_data
WHERE status != 'acquired' AND status != 'closed'
"""

Unknown = %sql $query_unknown

UnknownDF = pd.DataFrame(Unknown)

print("____________________________________________________________", "\n")

UnknownDF
```

```python
print("This is as compared to:", "\n")

SuccessDF.head()
```

```python
print('It looks like UnknownDF accidentally had "column 8" split into 2, forcing "columns 47, 48" to merge.')
```

```python
#Let's fix the UnknownDF

UnknownDF[8] = UnknownDF[8].astype(str) + UnknownDF[9].astype(str)

for col in range(9,47):
    UnknownDF[col] = UnknownDF[col+1]
    
for row in range(0,len(UnknownDF)):
    UnknownDF[48][row] = str(UnknownDF[48][row]).split(",")
    
for row in range(0,len(UnknownDF)):
    UnknownDF[47][row] = UnknownDF[48][row][0]
    
for row in range(0,len(UnknownDF)):
    UnknownDF[48][row] = UnknownDF[48][row][1]
    
UnknownDF
```

```python
print("We should check our tables carefully, now.","\n")

print("UnknownDF.head(3)","","","vs","","","SuccessDF.head(3)")

print("____________________________________________________________", "\n")

for col in range(0,49):
    print(list(UnknownDF[col].head(3)), list(SuccessDF[col].head(3)),"\n")
```

```python
#These look good enough to update our "Success" and "Fail" dataframes

for row in range(0,8):
    if UnknownDF[48][row] == "acquired":
        SuccessDF.loc[len(SuccessDF)+1] = UnknownDF.loc[row]
    else:
        FailDF.loc[len(FailDF)+1] = UnknownDF.loc[row]

print("Finally, we have accounted for", len(SuccessDF),"+",len(FailDF),"=",len(FailDF)+len(SuccessDF), "of", len(FullDF), "rows!")
```
