```python
import pandas as pd

query_full = """
SELECT *
FROM StartupData.dbo.startup_data
"""

Full = %sql $query_full

FullDF = pd.DataFrame(Full)

print("____________________________________________________________", "\n")

print("#rows =", len(FullDF), "", "#cols =", len(FullDF.columns), "\n")

FullDF.head()
```

```python
query_success = """
SELECT *
FROM StartupData.dbo.startup_data
WHERE status = 'acquired'
"""

query_fail = """
SELECT *
FROM StartupData.dbo.startup_data
WHERE status = 'closed'
"""

Success = %sql $query_success

Fail = %sql $query_fail

SuccessDF = pd.DataFrame(Success)

FailDF = pd.DataFrame(Fail)
```

```python
query_unknown = """
SELECT *
FROM StartupData.dbo.startup_data
WHERE status != 'acquired' AND status != 'closed'
"""

Unknown = %sql $query_unknown

UnknownDF = pd.DataFrame(Unknown)
```

```python
UnknownDF[8] = UnknownDF[8].astype(str) + UnknownDF[9].astype(str)

for col in range(9,47):
    UnknownDF[col] = UnknownDF[col+1]
    
for row in range(0,len(UnknownDF)):
    UnknownDF[48][row] = str(UnknownDF[48][row]).split(",")
    
for row in range(0,len(UnknownDF)):
    UnknownDF[47][row] = UnknownDF[48][row][0]
    
for row in range(0,len(UnknownDF)):
    UnknownDF[48][row] = UnknownDF[48][row][1]
```

```python
for row in range(0,8):
    if UnknownDF[48][row] == "acquired":
        SuccessDF.loc[len(SuccessDF)+1] = UnknownDF.loc[row]
    else:
        FailDF.loc[len(FailDF)+1] = UnknownDF.loc[row]

print("Finally, we have accounted for", len(SuccessDF),"+",len(FailDF),"=",len(FailDF)+len(SuccessDF), "of", len(FullDF), "rows!")
```

```python
print("First of all, some of these fella's had angel investors!  Did this make a difference?", "\n")

print("Success:", SuccessDF[41].astype(int).sum(), "/", len(SuccessDF), "=", round(((SuccessDF[41].astype(int).sum())/len(SuccessDF))*100,4), "%", "\n")

print("Fail:", FailDF[41].astype(int).sum(), "/", len(FailDF), "=", round(((FailDF[41].astype(int).sum())/len(FailDF))*100,4), "%", "\n")

print("Wow, it seems success is likely determined by the startup administration/type/etc., not exactly by assured money.")
```

```python
#Let's make sure each state is listed in both the Success and Fail dataframes

TempSuccessDF = SuccessDF.groupby(1).count()[0]

TempFailDF = FailDF.groupby(1).count()[0]

TempFullDF = FullDF.groupby(1).count()[0]

position_a = 97

for A in range(position_a,position_a+26):
    for B in range(position_a,position_a+26):
        if chr(A).upper()+chr(B).upper() not in TempFullDF:
            pass
        else:
            if chr(A).upper()+chr(B).upper() in TempSuccessDF:
                pass
            else:
                TempSuccessDF[chr(A).upper()+chr(B).upper()] = 0
            if (chr(A).upper()+chr(B).upper() in TempFailDF) == True:
                pass
            else:
                TempFailDF[chr(A).upper()+chr(B).upper()] = 0
                
TempRatioDF = TempFullDF

for A in range(position_a,position_a+26):
    for B in range(position_a,position_a+26):
        if chr(A).upper()+chr(B).upper() not in TempFullDF:
            pass
        else:
            TempRatioDF[chr(A).upper()+chr(B).upper()] = str(round((TempSuccessDF[chr(A).upper()+chr(B).upper()] / TempFullDF[chr(A).upper()+chr(B).upper()] ) * 100, 4)) + "%"
            
print("Success: \n", TempSuccessDF, "\n")
print("Fail: \n", TempFailDF, "\n")
print("Ratio Sucess/Total: \n", TempRatioDF)
```
